Agora, é importante notar que todas nossas informações estão separadas em diferentes tabelas, mas para uma melhor visualização é interessante fazer uma agregação aqui. Eu sei que esse processo é mais comum na engenharia de variaveis, mas mesmo assim vou fazer aqui na analise exploratória mesmo. Para visualizar melhor, pois a natureza desses dados (serem fragmentados) exige isso.

```python
#Código abaixo feito com o auxílio de IA

# Vamos fazer um LEFT JOIN (ou inner join) usando o 'shape_id' como chave.
# O 'inner' garante que só manteremos as viagens que têm um desenho correspondente na tabela de shapes.
df_trips_com_geometria = pd.merge(
    df_trips_clean,
    df_shapes_geom_clean[['shape_id', 'shape', 'shape_distance']], # Trazemos só as colunas úteis do shapes
    on='shape_id',
    how='inner'
)

print(f"Total de viagens com geometria mapeada: {df_trips_com_geometria.shape[0]}")
display(df_trips_com_geometria.head())
```

```python
df_trips_com_geometria.info()
```

O número de linhas explodiu, se tornou 5844184, o que é muita coisa. Aparentemente o motivo disso é que provavelmente tem rotas duplicadas, tipo, teve essa mesma rota em varios dias diferentes na semana. A informação de quantas vezes uma rota ocorreu em uma semana ou em um dado período de tempo não é uma informação tão útil para mim, então vou só remover as repetidas mesmo. Se em algum momento eu for precisar disso, já sei que foi aqui que eu removi.

```python
df_rotas_unicas = df_trips_clean[['route_id', 'shape_id', 'trip_headsign', 'direction_id']].drop_duplicates()

df_rotas_unicas.shape[0]

#Agora sim, cruzamos as rotas ÚNICAS com o desenho da rota (shapes)
df_rotas_geom = pd.merge(
    df_rotas_unicas,
    df_shapes_geom_clean[['shape_id', 'shape', 'shape_distance']],
    on='shape_id',
    how='inner'
)

print("Total de rotas únicas com geometria mapeada:" + str(df_rotas_geom.shape[0]))
```

O que está evidente agora é que os nossos dados de transporte são só referentes ao Rio de Janeiro mesmo, enquanto isso os nossos dados de violência englobam outras areas, como Niterói por exemplo. Então vamos aplicar mais um filtro.

```python
print(f"Total de incidentes antes do filtro: {len(df_fc_clean)}")

# 1. Filtrando apenas a cidade do Rio de Janeiro
# Usamos o .copy() para evitar aquele aviso chato de 'SettingWithCopyWarning' do Pandas
df_fc_rio = df_fc_clean[df_fc_clean['city'] == 'Rio de Janeiro'].copy()

print(f"Total de incidentes após filtrar por Rio de Janeiro: {len(df_fc_rio)}")

# 2. Sanity Check: Verificando se sobrou alguma outra cidade
print("\nCidades restantes na base:")
print(df_fc_rio['city'].unique())

# 3. Atualizando o GeoDataFrame para as próximas análises espaciais
import geopandas as gpd
gdf_crimes_rio = gpd.GeoDataFrame(
    df_fc_rio,
    geometry=gpd.points_from_xy(df_fc_rio.longitude, df_fc_rio.latitude),
    crs="EPSG:4326"
)

print("\nBase filtrada e pronta para o Ranking de Risco!")
```

```python
fig, ax = plt.subplots(figsize=(14, 10))

# Camada 1: Rotas de ônibus (em laranja/amarelo)
# Certifique-se de que a variável gdf_rotas foi criada nos passos anteriores
if 'gdf_rotas' in locals() or 'gdf_rotas' in globals():
    gdf_rotas.plot(ax=ax, linewidth=0.4, alpha=0.3, color='darkorange', label='Rotas de Ônibus')
else:
    print("Atenção: A variável gdf_rotas não está carregada na memória. Execute o passo anterior de geometria.")

# Camada 2: Incidentes do Fogo Cruzado (em vermelho)
ax.scatter(df_fc_rio['longitude'], df_fc_rio['latitude'], #mudei só essa linha
           s=15, alpha=0.7, color='red', edgecolor='black', label='Tiroteios / Incidentes')

plt.title("Risco Operacional: Rotas de Ônibus vs. Áreas de Risco (RJ)", fontsize=16)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(markerscale=2)
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

Aaaaah agora sim! Vamos dar uma olhadinha nas rotas mais perigosas levando em consideração tiroteios até 300 metros perto dela.

```python
# O segredo é definir o crs="EPSG:4326" logo na criação
print("--- CALCULANDO O RANKING DE RISCO FINAL ---")

# 1. Garantindo que as rotas têm o "carimbo" de coordenadas GPS (EPSG:4326)
# (Isso conserta o erro de 'naive geometry')
gdf_rotas.crs = "EPSG:4326"

# 2. Convertendo para metros (SIRGAS 2000 / UTM 23S)
gdf_crimes_metric = gdf_crimes_rio.to_crs(epsg=31983)
gdf_rotas_metric = gdf_rotas.to_crs(epsg=31983)

# 3. Criando o Buffer de 300 metros e fazendo o Join
gdf_rotas_metric['buffer_300m'] = gdf_rotas_metric.geometry.buffer(300)
gdf_rotas_buffer = gdf_rotas_metric.set_geometry('buffer_300m')

join_crimes_rotas = gpd.sjoin(gdf_crimes_metric, gdf_rotas_buffer, predicate='within')

# 4. Gerando o Ranking
ranking_perigo = join_crimes_rotas.groupby(['route_id', 'trip_headsign']).size().sort_values(ascending=False).reset_index()
ranking_perigo.columns = ['Linha', 'Letreiro', 'Qtd_Tiroteios_Proximos']

print("\n--- AS 5 LINHAS MAIS EXPOSTAS (NOV/2023) ---")
display(ranking_perigo.head(100))
```
